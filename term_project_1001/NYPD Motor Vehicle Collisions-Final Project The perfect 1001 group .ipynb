```python
ZIP = np.repeat(ZIP_Code_NYC['zipcode'],len(DATEandTIME))
DATE = np.tile(DATEandTIME, len(ZIP_Code_NYC))
arrays = [ZIP,DATE]
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=['ZIPCode', 'DATEandTIME'])
```
